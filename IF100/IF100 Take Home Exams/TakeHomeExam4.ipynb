```python

# Take Home Exam 4

def choose_side():
    
  player_turn= input('Player1, please choose your side (either X or O): ')
  
  while player_turn != 'X' and player_turn != 'O':
    player_turn=input('Player1, please choose your side (either X or O): ')
  return player_turn
  
def display_envr(list):
    
  print(list[0] [0] + '|' + list[0] [1] + '|' + list[0] [2])
  print('-+-+-')
  print(list[1] [0] + '|' + list[1] [1] + '|' + list[1] [2])
  print('-+-+-')
  print(list[2] [0] + '|' + list[2] [1] + '|' + list[2] [2])
  
def make_choice(plyr):
    
  turn=input('Player' + plyr + ', please make your choice: ')
  
  turnList= turn.split(',')
  ret_list= []
  
  for value in turnList:
    ret_list.append(int(value))
  return ret_list
  
def imply_choice(newList, list, mark):
    
  if newList[list[0]] [list[1]] == ' ':
    newList[list[0]] [list[1]]= mark
    return newList
    
  else:
    print(newList[list[0]] [list[1]], ' is already in the location ', list[0], ',', list[1], '. Skipping this turn.', sep='')
    return newList
    
def check_board(list):
    
  for value in list:
      
   for value2 in value:
       
    if value2== ' ':
     return True
     
  return False
  
def isWinner(list):

  if list[0] [1] == list[0] [2] and list[0] [1] == list[0] [0] and (list[0] [0] == 'O' or list[0] [0] == 'X'):
    return list[0] [0]

  elif list[1] [0] == list[2] [0] and list[1] [0] == list[0] [0] and (list[0] [0] == 'O' or list[0] [0] == 'X' ):
    return list[0] [0]

  elif list[1] [1] == list[2] [2] and list[1] [1] == list[0] [0] and (list[0] [0] == 'O' or list[0] [0] == 'X' ):
    return list[0] [0]

  elif list[1] [1] == list[2] [1] and list[1] [1] == list[0] [1] and (list[0] [1] == 'O' or list[0] [1] == 'X' ):
    return list[0] [1]
    
  elif list[1] [1] == list[1] [2] and list[1] [1] == list[1] [0] and (list[1] [0] == 'O' or list[1] [0] == 'X' ):
    return list[0] [1]

  elif list[2] [2] == list[1] [2] and list[1] [2] == list[0] [2] and (list[0] [2] == 'O' or list[0] [2] == 'X' ):
    return list[0] [1]

  elif list[2] [0] == list[1] [1] and list[1] [1] == list[0] [2] and (list[0] [2] == 'O' or list[0] [2] == 'X' ):
    return list[0] [2]
    
  elif list[2] [1] == list[2] [2] and list[2] [1] == list[2] [0] and (list[2] [0] == 'O' or list[2] [0] == 'X' ):
    return list[2] [0]  
    
  else:
    return ' '
  
  
# Main Part

print('Welcome to the Tic-Tac-Toe Game!')

plyr1= choose_side()

if plyr1== 'X':
  plyr2= 'O'
  
else:
  plyr2='X'

location=[[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
display_envr(location)

while True:
    
  plyr1Choice= make_choice('1')
  location= imply_choice(location, plyr1Choice, plyr1)
  display_envr(location)
  
  if isWinner(location)== plyr1 or check_board(location)== False:
   break
   
  plyr2Choice= make_choice('2')
  location= imply_choice(location, plyr2Choice, plyr2)
  display_envr(location)
  
  if isWinner(location)== plyr2 or check_board(location)== False:
   break

if check_board(location)== False:
  print('The game is finished! It is a tie.')
  
else:
    
  if plyr1== isWinner(location):
   print('Player1 has won the game!')
   
  else:
   print('Player2 has won the game!')
```
