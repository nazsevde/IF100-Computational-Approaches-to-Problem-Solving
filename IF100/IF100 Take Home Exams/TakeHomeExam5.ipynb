```python
# open and read file
inFile=open("songs.txt")
content=inFile.readlines()

# input function
genre=str(input("Please enter the genre that you want to listen: ")).lower()
genresList=[]

for x in content:
  itemList = x.split(" | ")
  genreForm = itemList[1]
  
  if genreForm not in genresList:
    genresList.append(genreForm)
    
  else:
    continue

# make list
songNamesList=[]
popularityScoresList=[]
beatsScoresList=[]

# while function
while x in content:
    
  x = x.strip("*, ")
  itemList = x.split(" | ")
  
  songNamesList.append(itemList[0])
  itemList[2] = int(itemList[2])
  popularityScoresList.append(itemList[2])
  itemList[3] = int(itemList[3])
  beatsScoresList.append(itemList[3])
  
  inFile.close()

# control function
control=True

while control:
    
  if genre=='quit':
    print("The program is closed. ")
    break

  elif genre not in genresList:
    print(f'There is no such genre {genre}. ' , sep='')
    genre = input('Please enter the genre that you want to listen: ')
    continue

  else:
    songNames=[]
    popularityScores=[]
    beatsPerminuteScores=[]
    
    
    for x in content:
      itemList = x.split(" | ")
      
      if genre==itemList[1]:
        songName = str(itemList[0])
        popularityScore = int(itemList[2])
        beatsScore = int(itemList[3])
        
        songNames.append(songName)
        popularityScores.append(popularityScore)
        beatsPerminuteScores.append(beatsScore)
        
        
    bestPopular = max(popularityScores)
    bestPopular = int(bestPopular)

    if popularityScores.count(bestPopular)!=1:
        
      myList=[]
      
      for x in range(len(popularityScores)):
        if popularityScores[x]==bestPopular:
          myList.append(int(beatsPerminuteScores[x]))
          
      bestBeat = max(myList)
      idx = beatsPerminuteScores.index(bestBeat)
      keySongName = songNames[idx]
      
    else:
      idxBeats = popularityScores.index(bestPopular)
      keySongName = songNames[idxBeats]
      bestBeat = beatsPerminuteScores[idxBeats]
      
    print(f'You can listen {keySongName[2:]} that has {bestPopular} popularity and {bestBeat} beats per minute. ' , sep='')
    # input repetition
    genre = str(input('Please enter the genre that you want to listen: '))
```
