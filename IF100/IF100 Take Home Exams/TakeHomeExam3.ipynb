```python
data = str(input("Please enter the dataset you have: "))
food = str(input("PLease enter the food(s) you ate during the day: "))
calorie_amount = int(input("Please enter the amount of calories you can take in a day: "))
willing_food = str(input("Please enter the food you want to eat: "))

dataset = data.replace(",",":")
dataset = data.replace("&",":")
dataset = data.split(",")
food_group = food.split(",")

food_list = []
group_list = []
calorie_list = []
check_point = True

for item in dataset:
    
  calorie_idx = item.split(":")
  food_list.append(calorie_idx[0].capitalize())
  group_list.append(calorie_idx[1])
  calorie_idx1 = calorie_idx[2]
  calorie_idx2 = calorie_idx1.find("cals")
  calorie_idx3 = calorie_idx1[:calorie_idx2]
  calorie_list.append(calorie_idx3)

count = 0

for foods in food_group:
    
  if foods.capitalize() in food_list:
    calorie_idx4 = food_list.index(foods)
    calorie_idx5 = calorie_list[calorie_idx4]
    count = int(calorie_idx5) + int(count)
    
  else:
    print(foods.capitalize(), "is not in the dataset.")
    check_point = False
    
if willing_food.capitalize() in food_list and check_point:
  willing_food_idx1 = food_list.index(willing_food)
  calorie_idx6 = calorie_list[willing_food_idx1]
  total_calorie = count + int(calorie_idx6)
  
  if total_calorie >= calorie_amount:
    willing_food_group = group_list[willing_food_idx1]
    
    for item in dataset:
      calorie_idx = item.split(":")
      
      if calorie_idx[1] == willing_food_group:
        suggested_food = calorie_idx[0]
        suggested_food_idx1 = food_list.index(suggested_food)
        total_calorie_idx1 = calorie_list[suggested_food_idx1]
        other_calorie = count + int(total_calorie_idx1)
        
        if calorie_amount >= other_calorie:
          print("You cannot eat "+willing_food.capitalize()+". However, we are suggesting you to eat "+suggested_food.capitalize()+".")
          check_point = False
          break
      
    if check_point:
      print("There is no food you can eat from", willing_food_group, "group.")
      check_point = False
      
  elif check_point:
    print("You can eat", willing_food.capitalize(), "that you will get", total_calorie, "calories.")
    check_point = False
    
elif check_point:
  print(willing_food.capitalize(), "does not exist in the dataset.")
```
